<h2 dir=rtl align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
ูุณูุช ฑ: ุงูุชุฎุงุจ centroid ูุง
</font>
</h2>

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
ุฏุฑ ุงู ูุณูุช ุงุจุชุฏุง ุจุง ุงุณุชูุงุฏู ุงุฒ <code>np.random</code> ูุฑฺฉุฒ ุซูู ูุง ุฑุง ุจุฏูู ุฌุงฺฏุฐุงุฑ ุงูุชุฎุงุจ ฺฉูุฏ.
</font>

```python
def initialize_centroids(X, k):
    return X[np.random.choice(X.shape[0], k, replace=False)]
```

<h2 dir=rtl align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
ูุณูุช ฒ: ูุญุงุณุจู ูุงุตูู ููุงุท ุงุฒ centroid ูุง.
</font>
</h2>

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
ุฏุฑ ุงู ูุณูุช ุจุงุฏ ูุงุตูู ููู ููุงุท ุฑุง ุจุง ุชฺฉ ุชฺฉ ูุฑฺฉุฒ ูุง ุซูู ูุญุงุณุจู ฺฉูุฏ ู ุขู ุฑุง ุจุฑฺฏุฑุฏุงูุฏ.
</font>
</p>

```python
def compute_distance(X, centroids):
  return np.sqrt(((X[:, np.newaxis] - centroids) ** 2).sum(axis=2))
```

<h2 dir=rtl align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
ูุณูุช ณ: ูุญุงุณุจู ุงูุฏุณ ูุฒุฏฺฉ ุชุฑู centroid.
</font>
</h2>

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
ุฏุฑ ุงูุฌุง ุจุงุฏ ุจุฑุง ูุฑฺฉุฏุงู ุงุฒ ุฏุงุฏู ูุง ูุฒุฏฺฉ ุชุฑู centroid ุฑุง ุจุง ุชูุฌู ุจู ุฎุฑูุฌ ุชุงุจุน ูุงุตูู ูพุฏุง ฺฉูุฏ ู ุงูุฏุณ ุขู centroid ุฑุง ุจุฑฺฏุฑุฏุงูุฏ.
</font>
</p>

```python
def assign_clusters(X, centroids):
    return np.argmin(compute_distance(X, centroids), axis=1)
```

<h2 dir=rtl align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
ูุณูุช ด: Update ูุฑุงฺฉุฒ ุซูู
</font>
</h2>

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
ุฏุฑ ุงูุฌุง ูุฑฺฉุฏุงู ุงุฒ centroid ูุง ุฑุง ุจุง ูุงูฺฏู ฺฏุฑูุชู ุงุฒ ููุงุท ุฏุฑูู ฺฉ cluster ุจุฑูุฒุฑุณุงู ูโฺฉูู.
</font>
</p>

```python
def update_centroids(X, labels, k):
    new_centroids = np.zeros((k, X.shape[1]))
    for i in range(k):
        points_in_cluster = X[labels == i]
        if points_in_cluster.shape[0] > 0:
            new_centroids[i] = points_in_cluster.mean(axis=0)

    return new_centroids
```

<h2 dir=rtl align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
ูุณูุช ต: ุญููู Training
</font>
</h2>

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
ุฏุฑ ุงู ูุณูุช ูุฑุงุญู ฺฉู ุฏุฑ ุชูุถุญ ุงูฺฏูุฑุชู ฺฏูุชู ุดุฏ ุฑุง ุจุตูุฑุช ูพูุณุชู ุงูุฌุงู ูโุฏูู ุชุง ุฏุฑ ููุงุช ุชุบุฑ ฺฉู ุฏุฑ centroid ูุง ุฏุฑ ูุฑ ูุฑุญูู ุงุฌุงุฏ ูโุดูุฏ ุญุฏุงฺฉุซุฑ ุจู ุงูุฏุงุฒู <code>tol</code> ุจุงุดุฏ.
</font>
</p>

```python
def kmeans(X, k, max_iters=100, tol=1e-4):
    """K-Means clustering algorithm."""
    centroids = initialize_centroids(X, k)

    for i in range(max_iters):
        labels = assign_clusters(X, centroids)
        new_centroids = update_centroids(X, labels, k)
        if np.all(np.abs(new_centroids - centroids) < tol):
            break
        centroids = new_centroids
    return centroids, labels
```

```python
def segmentation(image , k = 3):
    reshaped_image = convert_to_2D(image)
    """
    Steps:
        1.Perforing K-Means
        2.Assigning the value of cluster centroid to each pixel and converting it to integer
        3.Reshaping the segmented image to the original 3D tensor.
    """
    centroids , labels = kmeans(reshaped_image , k)
    segmented_image = centroids[labels].astype(np.uint8)
    output = segmented_image.reshape(image.shape)

    return output
```
